```python
def create_cubes(n):
  cube = []
  for i in range(n+1):
    cubes = i**3
    cube.append(cubes)
  return cube
```

```python
for x in create_cubes(5):
  print(x)
```

```python
def create_cubes(n):
  for x in range(n+1):
    yield x**3
```

```python
for x in create_cubes(10):
  print(x)
```

```python
# using as a generator
def fibonacci(n):
  a = 1
  b = 1
  for x in range(n+1):
    yield a
    a,b = b,a+b
```

```python
for i in fibonacci(10):
  print(i)
```

```python
def fibonacci(n):
  output = []
  a = 1
  b = 1
  for x in range(n+1):
    output.append(a)
    a,b = b,a+b
  return output
```

```python
for x in fibonacci(10):
  print(x)
```

```python
def square(n):
    for i in range(n+1):
        yield i**2
```

```python
for x in square(10):
    print(x)
```

```python
def simple_generator():
    for x in range(3):
        yield x
```
