```python
# accesing functions outside their scope is called a decorator
```

```python
import time

def timer(func):
  def wrapper(*args,**kwargs):
    start = time.time()
    result = func(*args,**kwargs)
    end = time.time()
    print(f"{func.__name__} ran in {end-start} time ")
    return result
  return wrapper
```

```python
@timer
def example_time(n):
  time.sleep(n)
```

```python
example_time(1)
```

```python
# In the decorator function a function is defined
```

```python
def debug(func):
  def wrapper(*args,**kwargs):

    arg_value = ','.join(str(arg) for arg in args)
    kwargs_value = ','.join(f"{k}={v}" for k,v in kwargs.items())
    print(f"Calling {func.__name__}({arg_value},{kwargs_value})")
    result = func(*args,**kwargs)
    print(f"{func.__name__} returned {result}")
    return func(*args,**kwargs)
  return wrapper
```

```python
@debug
def greeting(name,greeting="hello"):
  return f"{greeting} {name}"

greeting("Abbas",'mehdi')
```

```python
def login_credentials(fx):
  def wrapper(username,password):
    username = ''
    while len(password) < 8 or len(password) > 10:
      password = input("Enter password = ")
    return fx(username,password)
  return wrapper
```

```python
@login_credentials
def login(username,password):

  print(f"Welcome {username}")
```

```python
import time

def cache(func):
  cache = {}
  def wrapper(*args,**kwargs):
    if args not in cache:
      cache[args] = func(*args,**kwargs)
    return cache[args]
  return wrapper
@cache
def long_running_function(a,b):
  time.sleep(4)
  return a + b
```

```python
long_running_function(1,2)
```
